```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

# Увеличиваем размер шрифта по умолчанию для всех элементов
plt.rcParams.update({'font.size': 36})

# Подготовка данных
years = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022])
revenue = np.array([83.1, 99.8, 105.5, 110.7, 115.6, 130.1, 155.4, 186.9, 230.8, 273, 320, 351.1, 365])
inflation = np.array([5.77, 7.42, 11.92, 8.39, 4.91, 3.05, 4.27, 2.52, 5.38, 12.91, 11.36, 6.45, 6.58])

# Расчет выручки с поправкой на инфляцию (в ценах 2010 года)
inflation_adjusted_revenue = revenue.copy()
for i in range(1, len(inflation)):
    inflation_adjusted_revenue[i] = inflation_adjusted_revenue[i] / np.prod(1 + inflation[:i]/100)

# Создаем модели для обычной выручки
X = years.reshape(-1, 1)
y_normal = revenue.reshape(-1, 1)
y_adjusted = inflation_adjusted_revenue.reshape(-1, 1)

# Линейная и полиномиальная регрессия для обычной выручки
linear_model = LinearRegression()
linear_model.fit(X, y_normal)

poly_features = PolynomialFeatures(degree=10)
X_poly = poly_features.fit_transform(X)
poly_model = LinearRegression()
poly_model.fit(X_poly, y_normal)

# Линейная и полиномиальная регрессия для скорректированной выручки
linear_model_adj = LinearRegression()
linear_model_adj.fit(X, y_adjusted)

poly_model_adj = LinearRegression()
poly_model_adj.fit(X_poly, y_adjusted)

# Прогноз на 2023 и 2024 годы
future_years = np.array([2023, 2024])
linear_predictions = linear_model.predict([[2023], [2024]])
poly_predictions = poly_model.predict(poly_features.transform([[2023], [2024]]))
linear_predictions_adj = linear_model_adj.predict([[2023], [2024]])
poly_predictions_adj = poly_model_adj.predict(poly_features.transform([[2023], [2024]]))

# Создаем два графика
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(24, 24))

# График 1: Номинальная выручка
ax1.scatter(years, revenue, color='blue', label='Фактические данные', s=200)
years_extended = np.append(years, future_years)
X_plot = years_extended.reshape(-1, 1)
linear_prediction = linear_model.predict(X_plot)
poly_prediction = poly_model.predict(poly_features.transform(X_plot))

ax1.plot(years_extended, linear_prediction, color='red', linewidth=3,
         label=f'Линейный прогноз\n(2023: {int(linear_predictions[0][0])})\n(2024: {int(linear_predictions[1][0])})')
ax1.plot(years_extended, poly_prediction, color='green', linewidth=3,
         label=f'Полиномиальный прогноз\n(2023: {int(poly_predictions[0][0])})\n(2024: {int(poly_predictions[1][0])})')

ax1.set_title('Прогноз номинальной выручки (млрд руб.)', fontsize=42, pad=20)
ax1.set_xlabel('Год', fontsize=36, labelpad=20)
ax1.set_ylabel('Выручка (млрд руб.)', fontsize=36, labelpad=20)
ax1.legend(fontsize=32, bbox_to_anchor=(1.15, 1))
ax1.grid(True)

# График 2: Выручка с поправкой на инфляцию
ax2.scatter(years, inflation_adjusted_revenue, color='blue', label='Фактические данные\n(с поправкой на инфляцию)', s=200)
linear_prediction_adj = linear_model_adj.predict(X_plot)
poly_prediction_adj = poly_model_adj.predict(poly_features.transform(X_plot))

ax2.plot(years_extended, linear_prediction_adj, color='red', linewidth=3,
         label=f'Линейный прогноз\n(2023: {int(linear_predictions_adj[0][0])})\n(2024: {int(linear_predictions_adj[1][0])})')
ax2.plot(years_extended, poly_prediction_adj, color='green', linewidth=3,
         label=f'Полиномиальный прогноз\n(2023: {int(poly_predictions_adj[0][0])})\n(2024: {int(poly_predictions_adj[1][0])})')

ax2.set_title('Прогноз выручки с поправкой на инфляцию (млрд руб. в ценах 2010)', fontsize=42, pad=20)
ax2.set_xlabel('Год', fontsize=36, labelpad=20)
ax2.set_ylabel('Выручка (млрд руб.)', fontsize=36, labelpad=20)
ax2.legend(fontsize=32, bbox_to_anchor=(1.15, 1))
ax2.grid(True)

plt.tight_layout()
plt.show()

# Вывод прогнозов и метрик качества моделей
print("\nПрогноз номинальной выручки (млрд руб.):")
print(f"Линейная регрессия:")
print(f"2023: {int(linear_predictions[0][0])} млрд руб.")
print(f"2024: {int(linear_predictions[1][0])} млрд руб.")
print(f"\nПолиномиальная регрессия:")
print(f"2023: {int(poly_predictions[0][0])} млрд руб.")
print(f"2024: {int(poly_predictions[1][0])} млрд руб.")

print("\nПрогноз выручки с поправкой на инфляцию (в ценах 2010 года):")
print(f"Линейная регрессия:")
print(f"2023: {int(linear_predictions_adj[0][0])} млрд руб.")
print(f"2024: {int(linear_predictions_adj[1][0])} млрд руб.")
print(f"\nПолиномиальная регрессия:")
print(f"2023: {int(poly_predictions_adj[0][0])} млрд руб.")
print(f"2024: {int(poly_predictions_adj[1][0])} млрд руб.")

# Расчет коэффициентов детерминации
linear_r2 = linear_model.score(X, y_normal)
poly_r2 = poly_model.score(X_poly, y_normal)
linear_r2_adj = linear_model_adj.score(X, y_adjusted)
poly_r2_adj = poly_model_adj.score(X_poly, y_adjusted)

print(f"\nКоэффициент детерминации (R²) для номинальной выручки:")
print(f"Линейная регрессия: {linear_r2:.4f}")
print(f"Полиномиальная регрессия: {poly_r2:.4f}")
print(f"\nКоэффициент детерминации (R²) для выручки с поправкой на инфляцию:")
print(f"Линейная регрессия: {linear_r2_adj:.4f}")
print(f"Полиномиальная регрессия: {poly_r2_adj:.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Увеличиваем размер шрифта
plt.rcParams.update({'font.size': 36})

# Исходные данные
years = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024])
stores = np.array([301, 330, 370, 390, 403, 478, 539, 638, 760, 852, 933, 990, 971, 972, 950])
areas = np.array([313, 347.3, 383.5, 397.8, 415.8, 484, 548.5, 637.2, 781.5, 900, 1014, 1099, 1085, 1188, 1222])
visitors = np.array([263, 291, 297, 307, 302, 304, 350, 407, 505, 589, 562, 636.5, 657.6, 701, 722])

# Известное количество сотрудников в 2019
employees_2019 = 42800

# Создаем оценки на основе разных метрик
employees_by_stores = (stores * employees_2019) / stores[9]  # 2019 год имеет индекс 9
employees_by_area = (areas * employees_2019) / areas[9]
employees_by_visitors = (visitors * employees_2019) / visitors[9]

# Взвешенная оценка (предполагаем, что площадь и количество магазинов более важны)
weighted_employees = (employees_by_stores * 0.4 + 
                     employees_by_area * 0.4 + 
                     employees_by_visitors * 0.2)

# Создаем график
plt.figure(figsize=(24, 12))

# Plotting
plt.plot(years, employees_by_stores, 'b--', linewidth=2, label='По количеству магазинов')
plt.plot(years, employees_by_area, 'g--', linewidth=2, label='По торговой площади')
plt.plot(years, employees_by_visitors, 'r--', linewidth=2, label='По посещаемости')
plt.plot(years, weighted_employees, 'k-', linewidth=4, label='Взвешенная оценка')

# Отмечаем известную точку 2019 года
plt.scatter([2019], [employees_2019], color='red', s=300, 
           label='Известное значение (2019)', zorder=5)

plt.title('Оценка количества сотрудников по годам', fontsize=42, pad=20)
plt.xlabel('Год', fontsize=36, labelpad=20)
plt.ylabel('Количество сотрудников', fontsize=36, labelpad=20)
plt.grid(True)
plt.legend(fontsize=32)

plt.tight_layout()
plt.show()

# Выводим результаты
print("\nОценка количества сотрудников по годам:")
print("\nГод    |  Взвешенная оценка  |  По магазинам  |  По площади  |  По посещениям")
print("-" * 75)
for i, year in enumerate(years):
    print(f"{year}  |  {int(weighted_employees[i]):6d}          |  {int(employees_by_stores[i]):6d}      |  {int(employees_by_area[i]):6d}    |  {int(employees_by_visitors[i]):6d}")

# Расчет некоторых метрик
avg_employees_per_store = weighted_employees / stores
avg_employees_per_area = weighted_employees / areas

print("\nКлючевые метрики:")
print(f"\nСреднее количество сотрудников на магазин:")
print(f"2019 (известное): {employees_2019/stores[9]:.1f}")
print(f"Среднее за весь период: {np.mean(avg_employees_per_store):.1f}")

print(f"\nСреднее количество сотрудников на 1000 кв.м:")
print(f"2019 (известное): {employees_2019/areas[9]*1000:.1f}")
print(f"Среднее за весь период: {np.mean(avg_employees_per_area)*1000:.1f}")

# Расчет темпов роста
annual_growth = np.diff(weighted_employees) / weighted_employees[:-1] * 100

print(f"\nСредний годовой прирост персонала: {np.mean(annual_growth):.1f}%")
print(f"Максимальный годовой прирост: {np.max(annual_growth):.1f}% ({years[np.argmax(annual_growth)+1]})")
print(f"Минимальный годовой прирост: {np.min(annual_growth):.1f}% ({years[np.argmin(annual_growth)+1]})")
```
